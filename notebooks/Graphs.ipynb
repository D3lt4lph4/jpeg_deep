```python
# Detection speed

ssd_300 = [(88, 74.0), (102, 74.0)]
ssd_dct = [(136, 60.0), (262, 60.0)]
ssd_dct_deconv = [(144, 53.5), (282, 53.5)]
ssd_dct_y = [(140, 59.8), (278, 59.8)]


ssd_resnet = [(88, 73.1), (108, 73.1)]
ssd_lcrfa = [(87, 70.7), (110, 70.7)]
ssd_lcrfat = [(98, 67.5), (176, 67.5)]
ssd_lcrfa_y = [(91, 71.0), (109, 71.0)]
ssd_lcrfat_y = [(101, 70.2), (174, 70.2)]
ssd_deconv_rfa = [(87, 68.8), (104, 68.8)]

texts = []

figure(num=None, figsize=(7, 5), dpi=80, facecolor='w', edgecolor='k')

# Draw a point at the location (3, 9) with size 1000
plt.scatter(*ssd_300[1], c="black", s=20)
texts.append(plt.text(ssd_300[1][0], ssd_300[1][1]+1, "SSD300"))
# plt.scatter(*ssd_300[1], c="grey", s=20)

plt.scatter(*ssd_dct[1], c="b", marker="*", s=20)
texts.append(plt.text(ssd_dct[1][0]-45, ssd_dct[1][1]-1, "SSD-DCT"))
# plt.scatter(*ssd_dct[1], c="cyan", marker="*", s=20)

plt.scatter(*ssd_dct_deconv[1], c="b", marker=">", s=20)
texts.append(plt.text(ssd_dct_deconv[1][0]-80, ssd_dct_deconv[1][1]+0.5, "SSD-DCT-Deconv*"))
# plt.scatter(*ssd_dct_deconv[1], c="cyan", marker=">", s=20)

plt.scatter(*ssd_dct_y[1], c="b", marker="X", s=20)
texts.append(plt.text(ssd_dct_y[1][0]+5, ssd_dct_y[1][1]+0.5, "SSD-DCT-Y*"))
# plt.scatter(*ssd_dct_y[1], c="cyan", marker="X", s=20)

plt.scatter(*ssd_resnet[1], c="black", marker="v", s=20)
texts.append(plt.text(ssd_resnet[1][0]+5, ssd_resnet[1][1]-0.8, "SSD300-Resnet*"))
# plt.scatter(*ssd_resnet[1], c="grey", marker="v", s=20)

plt.scatter(*ssd_lcrfa[1], c="r", marker="^", s=20)
texts.append(plt.text(ssd_lcrfa[1][0]+5, ssd_lcrfa[1][1]-1, "SSD-LCRFA*"))
# plt.scatter(*ssd_lcrfa[1], c="tomato", marker="^", s=20)

plt.scatter(*ssd_lcrfa_y[1], c="r", marker="^", s=20)
texts.append(plt.text(ssd_lcrfa_y[1][0]-60, ssd_lcrfa_y[1][1], "SSD-LCRFA-Y*"))
# plt.scatter(*ssd_lcrfa[1], c="tomato", marker="^", s=20)

plt.scatter(*ssd_lcrfat[1], c="r", marker="p", s=20)
texts.append(plt.text(ssd_lcrfat[1][0]+3, ssd_lcrfat[1][1]+0.5, "SSD-LCRFA-Thinner*"))
# plt.scatter(*ssd_lcrfat[1], c="tomato", marker="p", s=20)

plt.scatter(*ssd_lcrfat_y[1], c="r", marker="p", s=20)
texts.append(plt.text(ssd_lcrfat_y[1][0]+3, ssd_lcrfat_y[1][1]+0.5, "SSD-LCRFA-Y-Thinner*"))
# plt.scatter(*ssd_lcrfat[1], c="tomato", marker="p", s=20)

plt.scatter(*ssd_deconv_rfa[1], c="r", marker="D", s=20)
texts.append(plt.text(ssd_deconv_rfa[1][0]-72.8, ssd_deconv_rfa[1][1]-1.5, "SSD-Deconv-RFA*"))
# plt.scatter(*ssd_deconv_rfa[1], c="tomato", marker="D", s=20)

# adjust_text(texts)

# Set x axis label.
plt.xlabel("FPS", fontsize=10)

# Set y axis label.
plt.ylabel("mAP", fontsize=10)

# Set size of tick labels.
plt.tick_params(axis='both', which='major', labelsize=9)

plt.xlim(right=350) #xmax is your value
plt.xlim(left=0)

plt.ylim(top=85) #xmax is your value
plt.ylim(bottom=50)

plt.hlines(ssd_300[1][1], 0, 350, linestyles="dotted")
plt.vlines(ssd_300[1][0], 50, 100, linestyles="dotted")

red_patch = mpatches.Patch(color='red', label='Resnet based DCT networks')
blue_patch = mpatches.Patch(color='blue', label='VGG based DCT networks')
black_patch = mpatches.Patch(color='black', label='RGB Networks')
plt.legend(handles=[black_patch, blue_patch, red_patch], loc="upper left")

plt.savefig('map_fps.pdf')

# Display the plot in the matplotlib's viewer.
plt.show()

```
