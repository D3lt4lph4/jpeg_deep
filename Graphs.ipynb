```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Detection speed

ssd_300 = [(111, 74.0), (131, 74.0)]
ssd_resnet = [(114, 73.1), (151, 73.1)]
ssd_dct = [(132, 60.0), (256, 60.0)]
ssd_dct_deconv = [(149, 53.5), (283, 53.5)]
ssd_dct_y = [(147, 59.8), (268, 59.8)]
ssd_lcrfa = [(109, 70.7), (150, 70.7)]
ssd_lcrfat = [(109, 67.5), (186, 67.5)]
ssd_deconv_rfa = [(94, 68.8), (112, 68.8)]


# Draw a point at the location (3, 9) with size 1000
plt.scatter(*ssd_300[0], c="black", label="SSD300", s=20)
plt.scatter(*ssd_300[1], c="grey", s=20)

plt.scatter(*ssd_resnet[0], c="black", marker="v", label="SSD300-Resnet", s=20)
plt.scatter(*ssd_resnet[1], c="grey", marker="v", s=20)

plt.scatter(*ssd_dct[0], c="b", marker="*", label="SSD-DCT", s=20)
plt.scatter(*ssd_dct[1], c="cyan", marker="*", s=20)

plt.scatter(*ssd_dct_deconv[0], c="b", marker=">", label="SSD-DCT-Deconv", s=20)
plt.scatter(*ssd_dct_deconv[1], c="cyan", marker=">", s=20)

plt.scatter(*ssd_dct_y[0], c="b", marker="X", label="SSD-DCT-Y", s=20)
plt.scatter(*ssd_dct_y[1], c="cyan", marker="X", s=20)

plt.scatter(*ssd_lcrfa[0], c="r", marker="^", label="SSD-LCRFA", s=20)
plt.scatter(*ssd_lcrfa[1], c="tomato", marker="^", s=20)

plt.scatter(*ssd_lcrfat[0], c="r", marker="p", label="SSD-LCRFA-Thinner", s=20)
plt.scatter(*ssd_lcrfat[1], c="tomato", marker="p", s=20)

plt.scatter(*ssd_deconv_rfa[0], c="r", marker="D", label="SSD-Deconv-RFA", s=20)
plt.scatter(*ssd_deconv_rfa[1], c="tomato", marker="D", s=20)

# Set chart title.
plt.title("mAP versus FPS", fontsize=19)

# Set x axis label.
plt.xlabel("FPS", fontsize=10)

# Set y axis label.
plt.ylabel("mAP", fontsize=10)

# Set size of tick labels.
plt.tick_params(axis='both', which='major', labelsize=9)

plt.xlim(right=350) #xmax is your value
plt.xlim(left=0)

plt.ylim(top=100) #xmax is your value
plt.ylim(bottom=50)

plt.hlines(ssd_300[0][1], 0, 350, linestyles="dotted")
plt.vlines(ssd_300[0][0], 50, 100, linestyles="dotted")

plt.legend()

plt.savefig('foo.png')
plt.savefig('foo.pdf')
# Display the plot in the matplotlib's viewer.
plt.show()

```
